```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your data from the CSV file
data = pd.read_csv('actor_kill_counts.csv')

# Sort the data by 'Kill_Count' in descending order
data = data.sort_values(by='kill_Count', ascending=False)

# Create the horizontal bar chart
plt.figure(figsize=(10, 6))
plt.barh(data['Actor'], data['kill_Count'], color='skyblue')
plt.title('Deadliest Actors in Hollywood')
plt.xlabel('Kill Count')
plt.ylabel('Actor')
plt.tight_layout()

# Label each bar with the corresponding actor's name and kill count
for index, value in enumerate(data['kill_Count']):
    plt.text(value + 10, index, str(value), va='center', color='black', fontweight='bold')

# Show the plot
plt.show()
```
