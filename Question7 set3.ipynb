```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv('us-marriages-divorces-1867-2014.csv')


years = [1900, 1950, 2000]
marriages_per_capita = []
divorces_per_capita = []

for year in years:
    year_data = data[data['Year'] == year]
    marriages_per_capita.append(year_data['Marriages'].values[0])
    divorces_per_capita.append(year_data['Divorces'].values[0])


plt.figure(figsize=(10, 6))
plt.bar(years, marriages_per_capita, width=0.4, label='Marriages per Capita')
plt.bar([x + 0.4 for x in years], divorces_per_capita, width=0.4, label='Divorces per Capita')
plt.title('Number of Marriages and Divorces per Capita in the U.S. (1900, 1950, 2000)')
plt.xlabel('Year')
plt.ylabel('Number per Capita')
plt.xticks([x + 0.2 for x in years], years)
plt.legend()
plt.tight_layout()

plt.show()
```
