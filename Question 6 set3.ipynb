```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
data = pd.read_csv('us-marriages-divorces-1867-2014.csv')

# Assuming your CSV file has columns 'Year', 'Marriages_per_Capita', and 'Divorces_per_Capita'
years = data['Year']
marriages_per_capita = data['Marriages']
divorces_per_capita = data['Divorces']

# Create the line plot
plt.figure(figsize=(10, 6))
plt.plot(years, marriages_per_capita, label='Marriages ', marker='o')
plt.plot(years, divorces_per_capita, label='Divorces per', marker='o')
plt.title('Number of Marriages and Divorces per Capita in the U.S. (1867-2014)')
plt.xlabel('Year')
plt.ylabel('Number per Capita')
plt.legend()
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```
