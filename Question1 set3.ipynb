```python

import pandas as pd


url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
users = pd.read_csv(url, sep='|')




mean_age_per_occupation = users.groupby('occupation')['age'].mean()


def male_ratio(series):
    return (series == 'M').sum() / series.count()

male_ratio_per_occupation = users.groupby('occupation')['gender'].apply(male_ratio)
sorted_male_ratio_per_occupation = male_ratio_per_occupation.sort_values(ascending=False)


min_max_age_per_occupation = users.groupby('occupation')['age'].agg(['min', 'max'])


mean_age_per_combination = users.groupby(['occupation', 'gender'])['age'].mean()


def gender_percentage(series):
    total_count = series.count()
    return (series == 'M').sum() / total_count, (series == 'F').sum() / total_count

gender_percentage_per_occupation = users.groupby('occupation')['gender'].apply(gender_percentage)


print("Mean age per occupation:")
print(mean_age_per_occupation)

print("\nMale ratio per occupation (sorted):")
print(sorted_male_ratio_per_occupation)

print("\nMinimum and Maximum ages per occupation:")
print(min_max_age_per_occupation)

print("\nMean age per combination of occupation and sex:")
print(mean_age_per_combination)

print("\nPercentage of women and men per occupation:")
print(gender_percentage_per_occupation)
```
